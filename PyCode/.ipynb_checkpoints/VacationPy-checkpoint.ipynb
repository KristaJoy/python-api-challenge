```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
weather_df = pd.read_csv('../weather_data.csv')
weather_df.head(10)
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Create a heatmap
figure_layout = {'width': '600px', 'height': '440px'}
fig = gmaps.figure(layout=figure_layout)
heatmap_layer=gmaps.heatmap_layer(weather_df[['Lat', 'Lng']], weights=weather_df['Humidity'], 
                                  max_intensity=100, point_radius=9)
heatmap_layer.opacity = .75
fig.add_layer(heatmap_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
#Enter a few weather criteria to get to a list of ideal locations
temp_ideal = weather_df.loc[weather_df['Max Temp'] > 75]
humidity_ideal = temp_ideal.loc[temp_ideal['Humidity'] < 75]
wind_ideal = humidity_ideal.loc[humidity_ideal['Wind Speed'] < 5]
hotel_df = wind_ideal.loc[wind_ideal['Cloudiness'] < 20]
hotel_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df.insert(9, "Hotel Name", "")
```

```python
hotel_df.head()
```

```python
#Call the API

target_radius = 15000
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    "radius": target_radius,
    "type": "lodging",
    "key": g_key,
}

# use iterrows to iterate through pandas dataframe
for index, row in hotel_df.iterrows():
    # get lat and lng from dataframe
    lat_lon = str(row['Lat']) + "," + str(row['Lng'])

    # add keyword to params dict
    params['location'] = lat_lon

    # assemble url and make API request
    print(f"Retrieving Results for Index {index}: {lat_lon}.")
    response = requests.get(base_url, params=params).json()
    #pprint(response) 
    results = response['results']
    # extract results
    try:
        hotel_df.loc[index, 'Hotel Name'] = results[0]['name']
    except:
        print('no data')
```

```python
print(hotel_df.head())
```

```python
hotel_df = hotel_df.drop(294)
hotel_df
```

```python
 # NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
figure_layout = {'width': '600px', 'height': '440px'}
fig = gmaps.figure(layout=figure_layout)

heatmap_layer = gmaps.heatmap_layer(weather_df[['Lat', 'Lng']], weights=weather_df['Humidity'], 
                                  max_intensity=100, point_radius=9)

marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)

heatmap_layer.opacity = .75

fig.add_layer(heatmap_layer)
fig.add_layer(marker_layer)

# Display figure
fig
```
